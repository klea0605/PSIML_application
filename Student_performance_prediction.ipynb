Kategoricke kolone

```python
categorical = (
    df_bool
    .drop(columns = ['age', 'absences', 'Passed'])
    .columns
)
```

```python
#debilu nije poenta da dobijes 70 kolona kao debil
#nemoj da koristis one-hot encoding vise kad imas milion kategorija 
def get_correct_dummies(df, categories):
    df_cpy = df.copy()
    for attr in categories:
        dummy = pd.get_dummies(df_cpy[attr], columns = [attr], prefix = attr)
        without_last = dummy.drop(dummy.columns.values[-1], axis = 'columns')
        df_cpy = df_cpy.drop([attr], axis = 'columns')
        df_cpy = pd.concat([df_cpy, without_last], axis = 'columns')
    return df_cpy
        
```

```python
pd.reset_option('display.max_rows')
display(get_correct_dummies(df_bool, categorical))
```

```python
df_encoded = get_correct_dummies(df_bool, categorical)
df_labels = df_bool['Passed']
#display(df_labels.head())
#display(df_train.head())
```

Podela na train/test itd

```python
type(df_encoded)
```

```python
df_train = (
    df_encoded
    .groupby('Passed', group_keys = False)
    .apply(lambda x: x.sample(frac = 0.75, replace = False, random_state = 42))
)

type(df_train)
df_train['Passed'].value_counts()
```

```python
df_test = df_encoded.drop(index = df_train.index)
df_test['Passed'].value_counts()
```

```python
df_features = df_encoded.drop(columns = 'Passed')
df_labels = df_encoded['Passed']
```

```python
train_features = df_train.drop(columns = 'Passed')
train_labels = df_train['Passed']

test_features = df_test.drop(columns = 'Passed')
test_labels = df_test['Passed']
```

Decision tree

```python
from sklearn import tree
from sklearn.metrics import accuracy_score
```

```python
for max_depth in range(1, 20):
    t = tree.DecisionTreeClassifier(criterion = 'entropy', max_depth = max_depth)
    clf = t.fit(train_features, train_labels)
    predictions = clf.predict(test_features)
    print(f'Accuracy: {accuracy_score(test_labels, predictions):.2f}%')
```
